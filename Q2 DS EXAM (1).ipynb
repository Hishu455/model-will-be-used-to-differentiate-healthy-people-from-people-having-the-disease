MODELLING

```python
from sklearn.model_selection import train_test_split
```

```python
cleaned_data.info()
```

Splitting data into dependent and independent variable

```python
X = cleaned_data.drop(['status'], axis = 1)
y = cleaned_data['status']
```

Organizing data into Training and Test

```python

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 , random_state = 10)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

Building multiple models(at least 4) and finding out best out of them

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(random_state = 10)
random_forest_model.fit(X_train,y_train)
predict_train_data = random_forest_model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_accuracy = accuracy_score(predict_train_data,y_train)
```

```python
print("Training Data Accuracy:- ",training_data_accuracy*100)
```

```python
x_test = random_forest_model.predict(X_test)
```

```python
testing_data_accuracy = accuracy_score(x_test,y_test)
```

```python
print("Testing Data Accuracy: ",testing_data_accuracy*100)
```
